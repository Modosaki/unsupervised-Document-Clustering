```python
stocks = 'ECOL CNK CHFN BG AXE APA '
stocks = stocks.split()
data = fyf.download(stocks, '2017-01-01', end="2018-01-01")['Close']
data.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(data)
plt.legend(['ECOL', 'CNK' , 'CHFN', 'BG', 'AXE', 'APA'], loc=2)
```

```python
returns = pd.DataFrame()
for stock in data:
    if stock not in returns:
        returns[stock] = np.log(data[stock]).diff()
returns = returns[1:]        
returns.head()  
```

```python
returns.describe()
```

```python
returns.corr()
```

```python
returns.corr()['BG'].sort_values(ascending=False)
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(returns, figsize=(16,12), alpha=0.3)
```
